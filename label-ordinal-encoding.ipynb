```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("customer.csv")
df.sample(5)
```

```python
df=df.iloc[:,2:]
df.head()
```

```python
x=df.iloc[:,:2]
y=df.iloc[:,2:]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=0)
X_train.shape,X_test.shape
```

```python
#ordinal encoding
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])

oe.fit(X_train)         
X_train = oe.transform(X_train)
X_test  = oe.transform(X_test)
```

```python
oe.categories
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
LE.fit(Y_train)
Y_train=LE.transform(Y_train)
Y_test=LE.transform(Y_test)
```

```python
#array to DF
```

```python
X_train=pd.DataFrame(X_train,columns=['review','education'])
X_test=pd.DataFrame(X_test,columns=['review','education'])
Y_train=pd.DataFrame(Y_train,columns=['purchased'])
Y_test=pd.DataFrame(Y_test,columns=['purchased'])
```

```python
#logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(X_train,Y_train)
ans=LR.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
a=accuracy_score(ans,Y_test)
a
```
